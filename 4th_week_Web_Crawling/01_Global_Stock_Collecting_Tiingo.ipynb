```python
from tiingo import TiingoClient  #initialize your client.  #TiingoClient 패키지에서 tiingo 모듈을 가져옴 
import pandas as pd 
import keyring 
```

```python
# Tiingp API 접속환경 설정 

api_key = keyring.get_password('tiingo', 'charmingcity')
config = {}             # initialize the dictionary
config['session'] = True  #map the key 'session' to the value 'True'   #To reuse the same HTTP Session across API calls
config['api_key'] = api_key  #map the key 'goodbye' to the value 'api_key'    #저장해 둔 API_KEY값 가져오기 
# 사전 구조 config = {'session': True, 'api_key': 'api_key'}     
client = TiingoClient(config) #접속환경

# Now you can use TiingoClient to make your API calls.
```

```python
# 티커 정보 가져오기 

tickers = client.list_stock_tickers() #list_stcok_tickers() 메서드를 통해서 티커 정보를 가져올 수 있다.
tickers_df = pd.DataFrame.from_records(tickers) # pandas.DataFrame.from_records: Convert structured or record ndarray to DataFrame.list 형태를 Pandas DataFrame 형태로 변환
```

```python
#티커 metata 가져오기 

ticker_metadata = client.get_ticker_metadata("VNQ") #get_ticker_metadata() 메서드 내에 티커 입력
print(ticker_metadata)
```

```python
ticker_history = client.get_dataframe("VNQ",                   #get_dataframe() 메서드 내에 티커를 입력하면 close(종가), high(고가), low(저가), open(시가), volumne(거래량), 수정주가, divCash(현금배당), splitFactor(주식분할 조정계수) 
                                      startDate= '2010-01-01',
                                      endDate = '2023-05-25',
                                      frequency='daily')
ticker_history
```

```python
# 무료 버전은 다우지수 30개 기업에 해당 메서드를 제공하여 월트 디즈니(DIS)로 진행 
```

```python
# 매일 가치지표 (시가총액, 기업가치, PER, PBR)

fundamentals_daily = client.get_fundamentals_daily('DIS')         #get_fundamentals_daily() 메서드 내에 티커를 입력하면 시가총액, 기업가치, PER, PBR, PEG 정보가 JSON 형태로 받아지며, from_record() 메서드를 통해 데이터프레임 형태로 변경해 준다. 
fundamentals_daily_df = pd.DataFrame.from_records(fundamentals_daily) #JSON 형태로 온 데이터를 dataframe 형태로 전환 

fundamentals_daily_df  
```
