# 1. 자료의 구조: 자료형

### 1. 자료의 구조: 자료형 
###   1-1 숫자형: 정수형(int)와 실수형(float)
###   1-2 문자열 자료형(str)
###   1-3 배열 자료형: 리스트(list)와 튜플(Tuple)
###   1-4 집합 자료형(set)
###   1-5 사전 자료형(dictionary)
###   1-6 부울 자료형(boolean)

#### 자료형: 프로그래밍을 할 때 쓰이는 숫자, 문자열 등 자료 형태로 사용하는 것
#### 숫자 자료형으로는 int형과 float형이 있으며 숫자 자료형 간에는 사칙 연산과 제곱 연산 등이 가능.
#### 문자열 자료형으로 str형이 있으며 큰따옴표 혹은 작은따옴표로 감싸서 정의
#### 배열 자료형으로 list와 tuple이 있으며 두 자료형 모두 인덱싱과 슬라이싱이 가능 (리스트의 요소는 변경할 수 없고 튜플은 변경할 수 있음)
#### 집합 자료형으로 set형이 있으며 요소 간 순서가 없고 요소의 중복을 허용하지 않음
#### 사전 자료형인 dictionary는 키와 값의 쌍으로 구성된 자료형, 키를 기준으로 데이터를 관리할 수 있음
#### 부울 자료형인 boolean은 참과 거짓을 의미하는 데이터 타입, 참을 의미하는 true와 거짓을 의미하는 false 두 가지의 값을 가지고 있음



<center>타 입|<center>설 명|<center>예
:---|:---|:---
int|정수형 데이터|100
float|소숫점을 포함한 실수|10.25
str|문자열|'Real Estate Financial Technology'
list|리스트, 순서가 있는 배열, 수정/추가/삭제가 가능한 자료 구조|[1, 2, 3, 'a', 'b']
tuple|튜플, 순서가 있는 배열, 수정/추가/삭제가 불가능한 자료 구조|(1, 2, 3, 'a', 'b')
dict|사전, {key: value}로 구성되어있는 자료 구조|{'Math': 99, 'English': 88, 'Korean': 78}
set|집합, {key}로 구성되어있는 자료 구조|{'a', 'b', 'c'}
bool|참/거짓|True/False



##### 튜플과 리스트의 가장 큰 차이는 요솟값을 변화시킬 수 있는가 여부이다. 즉 리스트의 요솟값은 변화가 가능하고 튜플의 요솟값은 변화가 불가능하다.
##### 따라서 프로그램이 실행되는 동안 요솟값이 항상 변하지 않기를 바란다거나 값이 바뀔까 걱정하고 싶지 않다면 주저하지 말고 튜플을 사용해야 한다.

```python
# 사전 자료형 딕셔너리(dictionary)는 키(key)와 값(value)의 쌍으로 구성된 순서가 없는 자료형 
# 리스트와 튜플의 경우에는 위치를 나타내는 인덱스로만 데이터를 관리할 수 있는 반면, 딕셔너리는 키를 기준으로 데이터를 관리할 수 있음

X = {"A":1, "B":2, "C":3}
print(X["A"])


X["D"] = 4
X["B"] = 5
print(X)

#del 함수

del(X["A"])
print(X)


# 딕셔너리 구성 요소를 확인하는 데 사용하는 메서드로는 key(), values(), items()
# key()는 딕셔너리의 모든 키를, values()는 딕셔너리의 모든 값을, items() 딕셔너리의 모든 쌍을 반환

X = {1:"A", 2:"B", 3:"C"}
print(X.keys())
print(X.values())
print(X.items())
```

```python
# 집합 자료형은 집합을 나타내는 자료형으로, 요소 간 순서가 없고 요소의 중복을 허용하지 않는 자료형
l = [1, 2, 3, 2, 1]
t = (1, 2, 3, 2, 1)

s1 = {1, 2, 3, 2, 1}
s2 = set([1, 2, 3, 2, 1])

print(l)
print(t)
print(s1)
print(s2)
```

```python
#집합 자료형은 집합을 나타내는 자료형으로, 요소 간 순서가 없고 요소의 중복을 허용하지 않는 자료형

a = [1,2,3,4,5,3,4,2,1,2,4,5]

a_set = set(a)

print(a_set)
```

```python
# 집합 자료형에서는 교집합, 합집합, 차집합 연산을 지원

# 교집합 연산: &, intersection 메서드
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
print(s1 & s2)
print(s1.intersection(s2))

# 합집합 연산: |, union 메서드
print(s1|s2)
print(s1.union(s2))

#차집합 연산: -, difference 메서드
print (s1 - s2)
print(s1.difference(s2))
```

```python
# 교집합, 합집합


a = ['사과', '감', '배', '수박', '딸기']
b = ['배', '사과', '포도', '참외', '수박']

a_set = set(a)
b_set = set(b)

print(a_set & b_set)  #교집합 
print(a_set | b_set)  #합집합
```

###    <수행할 문장1>
###    <수행할 문장2>

```python
# while문은 조건을 만족하는 한 계속 while문 아래 있는 코드를 반복 실행 

x = 1               #x를 1로 초기화
while x <=5 :      #x가 5보다 작거나 같은지를 검사
    print(x)
    x +=1          #x를 1만큼 증가, X가 6으로 조건을 만족하지 않을 때까지 조건문을 반복 
```

#### for문과 while 문 비교

```python
# for문과 달리 while문의 경우 변수를 초기화 해야 하며 업데이트 조건도 설정해야 함 
# 특정한 요소를 순회하는 경우는 while문 보다 for문이 편리
# for문은 한 변수가 순회 가능한 객체를 순회하면서 코드 블록을 반복해서 실행하는 데 사용하며,range 함수를 통해 특정 횟수만큼 반복하는데 사용
# while문은 특정한 조건을 만족하는 동안 코드 블록을 반복해서 실행, 조건이 계속 참이면 무한히 코드를 실행(무한 루프)
# while문의 경우 코드의 가독성을 위해 의도적으로 무한 루프에 빠지게 하고 원하는 조건을 충족하면 break문을 통해 나올 수 있게 설정 할 수 있음

for i in range(1,6):
    print(i)
    
```

```python
# while 문은 [특정한 조건을 만족하는 코드를 반복]해서 수행해야 하는 상황에 적합

n = 1 
while 2 ** n < 10 ** 8: # 2의 제곱이 언제 1억보다 커지는지 확인 
    n += 1
print(n)
# n 값이 27일 때 10**8보다 커짐 
```

```python
# 함수의 매개변수 
# 매개변수는 함수에 입력으로 전달된 값을 받는 변수를 의미
# 인수는 함수를 호출할 때 전달되는 입력 값을 의미 

def cal(a,b):    #a, b는 매개변수 
    return a+2*b

result = cal(1,2)  # 1, 2는 인수 
print(result)
```

```python
# 함수

def hello():
    print('안녕!')
    print('좋은 날씨야')
    
hello()
```

```python
# 입력값의 개수를 지정하지 않고 모두 받는 방법
## 여러 개의 인수를 하나의 매개변수로 받을 때 관례적으로 'args'라는 이름을 사용 

def cal(*args):
    for name in args:
        print(f'{name}님 연락주세요')
              
cal('제명', '주연', '열매', '성빈', '지현')
```

```python
# 메서드,  메서드는 클래스 내부에 정의된 함수 

class Calulator:              #add_ver1과 add_ver2라는 메서드를 갖는 calculator라는 클래스를 정의
    def __int__(self, value):   #생성자, 생성자를 통해 인스턴스의 value의 속성을 정의 
        self.value = value
        
    def add_ver1(self, input_value): # 메서드 정의 1 
        self.value +=input_value     #input_value를 더해 value의 속성을 업데이트하는 메서드 add_ver1을 정의
        
    def add_ver2(self, input_value): # 메서드 정의 2 
        return self.value + input_value  #input_value를 더한 값을 반환하는 메서드 add_ver2를 정의
```

```python
# 같은 클래스를 통해서 만들어진 인스턴스라도 그 특성은 다를 수 있다. 
# 가령 L1 = [1, 2, 3]과 L2 = [5, 6]은 모두 리스트라는 클래스의 인스턴트이지만, 리스트 안에 속한 요소라는 특성은 서로 다름
# 이러한 객체의 특성을 파이썬에서는 속성(attribute)라고 부른다.
# 객체의 속성은 생성자(__int__)라는 함수를 통해 할당됨

class Stock:     # stock이라는 이름의 클래스 선언
    def __init__(self, name, market): # __init__라는 이름의 함수(생성자)를 생성, 클래스 내부에 있는 함수를 메서드라고 함
        self.name = name      # 속성 정의 1 
        self.market = market    # 속성 정의 2 
        
samsung = Stock("삼성전자", "코스피") #samsung 객체 생성, Stock 클래스 호출 
print(samsung.name, samsung.market)

#__init__ 메서드는 james = Person()처럼 클래스에 ( )(괄호)를 붙여서 인스턴스를 만들 때 호출되는 특별한 메서드입니다. 즉, __init__(initialize)이라는 이름 그대로 인스턴스(객체)를 초기화
```

```python
#클래스(class)

class Monster(): #class는 Monster 객체들을 만들어 내는 붕어빵 기계
    hp = 100
    alive = True

    # damage 메서드, 클래스 내부에 있는 함수를 메서드라고 함
    # 파이썬 메서드의 첫 번째 매개변수 이름은 관행적으로 self를 사용한다. 객체를 호출한 객체 자신이 전달되기 때문에 self라는 이름을 사용.
    def damage(self, attack):        #메서드의 매개변수(self, attack) 
        self.hp = self.hp - attack   #메서드의 수행문  
        if self.hp < 0:
            self.alive = False
 
    #status_check 메서드, 클래스 내부에 있는 함수를 메서드를 하고 함
    def status_check(self):   
        if self.alive:
            print('살아있다')
        else:
            print('죽었다')
```

```python
#m1는 클래스의 인스턴스
#m1 객체는 Monster 클래스의 인스턴스이다. 인스턴스는 특정 객체(m1)가 어떤 클래스인지 관계 위주로 설명할 때 사용
#m1은 객체(object).
m1 = Monster()    
m1.damage(120)  #m1 객체를 통해 damage 메소드를 호출, def(self, attack) 메서드의 매개변수인 self에 m1 객체가 자동으로 전달됨,
m1.status_check()
```

```python
#m2는 Mosnter 클래스의 인스턴스

m2 = Monster()
m2.damage(20)
m2.status_check()
```

### 생산자와 속성  

### 클래스의 상속
#### 클래스 A에서 상속된 클래스 B가 있다고 할때,
#### A를 부모(Parent) 클래스라고 하며,
#### B를 자식(Child) 클래스라 한다.

```python
# 부모(상위) 클래스
class Person:       
    def __init__(self,name,phone=None):  # 생성자
        self.name = name
        self.phone = phone
        
    def get_info(self):
        return self.name,self.phone
```

```python
# Person 클래스를 통해 p 인스턴스 생성

p = Person('홍길동','010-1234-1234')   # 인스턴스의 생성
print(p.get_info())
```

```python
print(Person.__bases__)  # base 클래스를 모두 tuple로 반환 
                         # 파이썬의 모든 클래스의 base 클래스는 object 이다.
```

```python
# 자식(하위) 클래스
class Employee(Person):                   # base 클래스는 괄호 안에 표현한
    def __init__(self,name,phone,position,salary):
#         self.name = name
#         sel.phone = phone
#         Person.__init__(slef,name,phone) # Person 클래스 생성자 호출
          super().__init__(name,phone)     #  상위 클래스 생성자 호출
          self.position = position
          self.salary  = salary
       

    # 메서드의 확장
    # 하위 클래스에서 그 속성을 변화시키지 위해서 상위 클래스의 메서드를 호출하고, 
    # 그 결과를 활용하는 것을 말한다.
    def get_info(self):
        ret = super().get_info()
        return ret, self.position, self.salary        
            
#  super() 메서드 : 현재 클래스의 상위(수퍼) 클래스를 얻어 낸다   
```

```python
m1 = Employee('김','1111','과장',500)
print(m1.get_info())

m2 = Employee('홍','2222','부장',500)
print(m2.get_info())
```
